```python
#Hello and welcome to my discounted cashflow code
#This code discounts cashflows using exponential growth


import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 40
#I'll assume the next 40 years will pay a cashflow
#After year 40 the cashflows become negligable
CF = np.asarray([1]*t)
#the cashflow pays 1 dollar a year
r=.1
#I will assume a discount rate of 10% anually
```

```python
def discount_cashflow(CF,t):
    #given t we estimate what a dollar would be worth at each time
    
    DCF=[]
    for i in range(t):
        #how much to discount at each year
        d = 1/(1+r)**i
        DCF.append(float(CF[i])*float(d))
    return DCF
```

```python
for i in range(t):
    (CF*(1.04)**i)*(1-r)**i
```

```python
#get the NPV of each year's cashflow
y = discount_cashflow(CF,t)
```

```python
#graph results
plt.plot(range(t),y,'go')
plt.ylabel('NPV of a dollar')
plt.xlabel('Years in the future')
plt.show()
```

```python
#After year 40 cashflows are worth pennies on the dollar
```

```python
#sum up the value of a dollar every year for 40 years
s = sum(y)
s
```

```python
#The DCF formula is PV = CF*(1-r)^t for each cashflow

#the multiplication happens after the exponent calculation
#we can multiply the exponent by the cashflow ourselves

#sum the value of ten dollars a year for 40 years
s*10
```

```python
#now we can apply our discounted cashflow model
#suppose a landlord earns $14000 each year as lease payment
#what is the NPV of this lease?
y = discount_cashflow([14000]*t,t)
#graph results
plt.plot(range(t),y,'go')
plt.ylabel('PV of a cashflow')
plt.xlabel('Years in the future')
plt.show()
```

```python
#the sum of these discounted cashflows
print('The NPV of the lease is $'+str(sum(y))[0:9])
```

```python
#This works well for constant cashflows
#what if cashflows are increasing?
```

```python
#Company ABC is expected to net 15M each year
#They are growing at a rate of 4% annually 

CF_ABC = [15]*t
#calculate cashflows over t years
def calculate_growth(CF,t):
    for year in range(t):
        #Cashflows grow at a rate of 4%
        CF_ABC[year] = float(CF_ABC[year]*((1.04)**year))
    
    #result
    return CF_ABC
```

```python
#a plot of increasing cashflows
plt.plot(calculate_growth(CF_ABC,t),'bo')
plt.ylabel('Projected Cashflows')
plt.xlabel('Years in the future')
plt.show()
```

```python
#get discounted cashflows
y = discount_cashflow(CF_ABC,t)
```

```python
#the discounted but growing cashflows
plt.plot(y,'ro')
plt.ylabel('Projected Cashflows')
plt.xlabel('Years in the future')
plt.show()
```

```python
#the sum of our discounted cash flows
sum(y)
```

```python
#another method used is the gordon growth model
def gordon_growth(CF1):
    #discounting with growth
    r=.1
    g = .04
    return CF1/(r-g)
```
